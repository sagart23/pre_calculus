# Quadratic Equations
* A quadratic equation is a polynomial equation of degree 2, which means it contains a term with a variable raised to the power of 2. It takes the form:
$$ a x^2 + bx + c = 0 $$
where, a, b, and c are coefficient and real numbers and also $ a \ne 0 $. <br>
If a is equal to 0 that equation is not valid quadratic equation.

```python
# required Python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve, sqrt
from IPython.display import display, Math
from ipywidgets import interact, FloatSlider, widgets
from IPython.display import HTML
```

```python
# Define the function
def quadeq(x):
    return a * x**2 + b * x + c


# Set coefficients
a = 1
b = -5
c = 6

# Generate x values
x = np.linspace(-10, 10, 21)

# Calculate corresponding y values
y = [quadeq(val) for val in x]

# Create a pandas DataFrame
data = pd.DataFrame({'x': x, 'y': y})
data['x'] = data['x'].apply(lambda x: round(x, 2))
data['y'] = data['y'].apply(lambda x: round(x, 2))
# Create styled DataFrame for highlighting

# Create the scatter plot
plt.scatter(x, y)

# Set labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Graph of  " f'$y = {a}x^2 + {b}x + {c}$')

# Option 2: Using Jupyter Notebook (automatic display)
print("Data Table:")
display(data)
#html_data = styled_data.to_html()
#print("Data Table:")
#print(html_data)


# Show the plot
plt.show()
```

## Roots of the equations
* $x_1 \text{ and } x_2$ are called roots of the equations if the equation take value zero when x takes these values.
* The quadratic equations can be formed if the roots are known:
  $$ x^2 - (\text{Sum of roots })x + (\text{Product of roots})$$
* Roots of the quadratic equations can also be calculated using the following formula
  $$ x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$
* By knowing the value of $\sqrt{b^2 - 4ac}$ the nature of the roots can be told.
* $\Delta = \sqrt{b^2 - 4ac}$  is called the discriminant.
* If $ \Delta < 0 $ then the roots are imaginary (or complex)
* If $ \Delta \ge 0 $ then the roots are real.
  * If $ \Delta > 0 $ then the roots are real and unequal.
  * If $ \Delta = 0 $ then the roots are real and equal.

```python
## Roots are real and unequal
x = np.linspace(-10, 10, 400)
y = x**2 - 2*x - 15

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"$x^2 - 5x + 6$")
plt.text(-5, 60, 'The curve cuts the x axis at two points \n each point is called a root of the equation', style='italic',
        bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 10})
plt.plot(-3,0, 'ro')
plt.annotate('root', xy=(-3,0), xytext=(-5,40),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=15)
plt.plot(5,0, 'ro')
plt.annotate('root', xy=(5,0), xytext=(6,40),
            arrowprops=dict(facecolor='black', shrink=0.05), fontsize=15)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Graph of the Quadratic Equation : Roots are real and unequal')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

## Graph of Quadratic Equations
* The code below takes input A,B and C
* Plots a graph and gives the roots of the equations.

```python
# Code below ask for input A, B and C and polts graph ond give solutions to the quadractic equations

# Input values for A, B, and C
A = float(input("Enter the value of A (not zero): "))
B = float(input("Enter the value of B: "))
C = float(input("Enter the value of C: "))

# Display the quadratic equation in LaTeX form
equation_latex = f"{A}x^2 + {B}x + {C} = 0"
display(Math(equation_latex))

# Define the quadratic function
def quadratic_function(x):
    return A*x**2 + B*x + C

# Calculate the vertex
vertex_x = -B / (2 * A)
vertex_y = quadratic_function(vertex_x)

# Display the vertex
vertex_latex = f"vertex: \\left( {round(vertex_x, 4)}, {round(vertex_y, 4)} \\right)"
display(Math(vertex_latex))

# Plot the quadratic function
x = np.linspace(-10, 10, 400)
y = quadratic_function(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f"${A}x^2 + {B}x + {C}$")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Graph of the Quadratic Equation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
# plot the vertex
plt.plot(vertex_x, vertex_y, 'bo')  # 'bo' means blue color and circle marker
#plt.annotate(f"Vertex ({vertex_x:.4f}, {vertex_y:.4f})", (vertex_x, vertex_y), textcoords="offset points", xytext=(0,10), ha='center')


# Calculate the discriminant (b^2 - 4ac)
discriminant = B**2 - 4*A*C

# Calculate the roots of the quadratic equation
x_symbol = symbols('x')
equation = Eq(A*x_symbol**2 + B*x_symbol + C, 0)
roots = solve(equation, x_symbol)

# Display the roots
print(f"The roots of the equation are: {roots}")

# Display the roots in the specified form, rounded to 4 decimal places
for i, root in enumerate(roots):
    rounded_root = round(root.evalf(), 4)
    root_latex = f"x_{i+1} = {rounded_root}"
    display(Math(root_latex))
    # Plot the roots (x1, 0) and (x2, 0) if they are real
    if root.is_real:
        plt.plot(float(root.evalf()), 0, 'ro')  # 'ro' means red color and circle marker

# Display the roots in the specified form
root_latex = f"x = \\frac{{{-B} \\pm \\sqrt{{{discriminant}}}}}{{2*{A}}}"
display(Math(root_latex))

plt.show()
```

```python
# Define the quadratic function
def quadratic_function(x, A, B, C):
    return A*x**2 + B*x + C

# Function to update the plot
def update_plot(A, B, C):
    # Display the quadratic equation in LaTeX form
    equation_latex = f"{np.round(A,2)}x^2 + {np.round(B,2)}x + {np.round(C,2)} = 0"
    display(Math(equation_latex))

    # Plot the quadratic function
    x = np.linspace(-5, 5, 200)
    y = quadratic_function(x, A, B, C)

    plt.figure(figsize=(10,6))
    plt.plot(x, y, label=f"${A}x^2 + {B}x + {C}$")
    #plt.ymax = 100
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title('Graph of the Quadratic Equation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

# Create sliders for A, B, and C
A_slider = FloatSlider(value=1.0, min=-5.0, max=5.0, step=1, description='A:')
B_slider = FloatSlider(value=0.0, min=-5.0, max=5.0, step=1, description='B:')
C_slider = FloatSlider(value=0.0, min=-5.0, max=5.0, step=1, description='C:')

# Use interact to create an interactive plot
interact(update_plot, A=A_slider, B=B_slider, C=C_slider)
```

```python
a = 1
b = -3
c = -28
equation_latex = f"y = {a}x^2 + {b}x + {c}, \quad -10 \le x \le 10"

slider = widgets.FloatSlider(
    description='$x$',
    min=-10,
    max=10,
    value=2
)
text = widgets.FloatText(disabled=True, description='$y$')

stored_values = []  # List to store (value, slider_value) tuples

def compute(*ignore):
    value = a*slider.value**2 + b*slider.value + c
    text.value = np.round(value, 4)

    # Store current value and slider value
    stored_values.append((value, slider.value))

slider.observe(compute, 'value')

# Code to plot the stored_values list (can be placed after the widget display)

def plot_values():
    """
    Plots the values stored in the 'stored_values' list.
    """

    x_values, y_values = zip(*stored_values)  # Unpack stored_values into separate lists
    plt.plot(y_values, x_values)
    plt.xlabel('Slider Value (x)')
    plt.ylabel('Calculated Value (y)')
    plt.title('Plot of Stored Values')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.show()

# Display the widgets
display(Math(equation_latex))
display(widgets.VBox([slider, text]))

```
